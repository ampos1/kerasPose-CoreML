```python
def data_processing(df):
    
    check = ['a','e','i','o','u'] 
    
    for i in range(0,df['values'].count()):
        
        if type(df['values'][i]) != str:
            val = df['values'][i]
            df['values'][i] = df['values'][i].to_string()

        if 'real' in df['values'][i]:
            temp = re.search('>(.*)</',df['values'][i])
            df['values'][i] = temp.group(1)
    
        for vowel in check:
            if vowel in list(df['values'][i]):
                df.drop(index = i,inplace = True)
                break
            
    return df

def update_index(df):
    df = df.reset_index()
    df.drop('index',axis=1,inplace=True)
    return df
```

```python
def extract_x_values(data):
    temp_arr = pd.DataFrame(np.zeros((int(data.size/4),1)))
    for i in range(0,int(data.count()),4):
        if(i == 0):
            temp_arr[0][i] = data['values'][i]
        else:
            temp_arr[0][(i/4)] = data['values'][i]
    return temp_arr
```

```python
def extract_y_values(data):
    temp_arr = pd.DataFrame(np.zeros((int(data.size/4),1)))
    for i in range(1,int(data.count()),4):
        if(i == 1):
            temp_arr[0][i-1] = data['values'][i]
        else:
            temp_arr[0][(i-1)/4] = data['values'][i]
    return temp_arr
```

```python
def extract_z_values(data):
    temp_arr = pd.DataFrame(np.zeros((int(data.size/4),1)))
    for i in range(2,int(data.count()),4):
        if(i == 2):
            temp_arr[0][i-2] = data['values'][i]
        else:
            temp_arr[0][(i-2)/4] = data['values'][i]
    return temp_arr
```

```python
def extract_c_values(data):
    temp_arr = pd.DataFrame(np.zeros((int(data.size/4),1)))
    for i in range(3,int(data.count()),4):
        if(i == 3):
            temp_arr[0][i-3] = data['values'][i]
        else:
            temp_arr[0][int((i-3)/4)] = data['values'][i]
    return temp_arr
```

```python
def restructure(data):
    organized_df = pd.DataFrame()
    for i in range(0,int(data.shape[0]/16)):
        lower_bound = 0
        upper_bound = 15
        curr_lower = lower_bound + (i*16)
        curr_upper = upper_bound + (i*16)
        
        df_splice = data.loc[curr_lower:curr_upper,:]
        df_splice.reset_index(inplace=True,drop=True)
        df_splice = df_splice.transpose()
        organized_df = pd.concat([organized_df,df_splice])
    
    return organized_df
```
