```python
import pandas as pd
import numpy as np
```

```python
address = pd.read_csv("address.csv",index_col=0)
backswing = pd.read_csv("backswing.csv",index_col=0)
impact = pd.read_csv("impact.csv",index_col=0)
finish = pd.read_csv("finish.csv",index_col=0)
```

```python
address['Address Position'] = np.ones((address.count()[0],))
backswing['Backswing Position'] = np.full((backswing.count()[0],),2)
impact['Impact Position'] = np.full((impact.count()[0],),3)
finish['Finish Position'] = np.full((finish.count()[0],),4)
```

```python
df = pd.concat([address,backswing,impact,finish],sort=False)
df.fillna(value=0,inplace=True)
```

```python
data = df.loc[:,'0':'15']
labels = df.loc[:,'Address Position':'Finish Position']
```

```python
index_val_arr = data.index.values
trial_number = np.array((range(0,int(data.count()[0]/4))))
arrays = np.array([np.repeat(trial_number,4),index_val_arr])
multi = pd.MultiIndex.from_arrays(arrays,names=('Trial','Coordinate'))
```

```python
df = pd.DataFrame(np.concatenate((address,backswing,impact,finish),axis=0),index=multi)
df = df.reindex(multi)
```

```python
one_hot = pd.get_dummies(df.loc[:,16]).rename(columns= {1.0:"Address Position", 2.0:"Backswing Position",
                                                            3.0:"Impact Position",4.0:"Finish Position"})
```

```python
complete_df = pd.concat([df,one_hot],axis=1)
complete_df.drop(columns=16,inplace=True)
```
