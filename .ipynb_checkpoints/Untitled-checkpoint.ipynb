```python
import pandas as pd
import keras as keras
import numpy as np
```

```python
address = pd.read_csv("address_data_csv",index_col=0)
impact = pd.read_csv("impact_data_csv",index_col=0)
finish = pd.read_csv("finish_data_csv",index_col=0)
```

```python
address['Address Position'] = np.ones((819,))
impact['Impact Position'] = np.full((819,),2)
finish['Finish Position'] = np.full((819,),3)
```

```python
temp = pd.concat([address,impact,finish],sort=False)
temp.fillna(value=0,inplace=True)
```

```python
data = temp.loc[:,'0':'15']
labels = temp.loc[:,'Address Position':'Finish Position']
```

```python
index_val_arr = data.index.values
trial_number = np.array((range(0,819)))
arrays = np.array([np.repeat(trial_number,3),index_val_arr])
```

```python
multi = pd.MultiIndex.from_arrays(arrays,names=('Trial','Coordinate'))
```

```python
df = pd.DataFrame(np.concatenate((address,impact,finish),axis=0),index=multi)
```

```python
new_df = df.reindex(multi)
```

```python
one_hot = pd.get_dummies(new_df.loc[:,16]).rename(columns= {1.0:"Address Position",2.0:"Impact Position",3.0:"Finish Position"})
```

```python
new_df = pd.concat([new_df,one_hot],axis=1)
new_df.drop(columns=16,inplace=True)
```
